# 4. WRANGLING

```python
def cat_num_type(df, max_cat = 20):
    for col in tqdm(df.columns):
        if df[col].nunique()<max_cat:
            df[col] = df[col].astype(str)
        else:
            df[col] = df[col].astype(float)
    return df
```

```python
df_train.drop(columns = ['cliente'], inplace = True)
```

```python
df_train = cat_num_type(df_train)
```

```python
def numeric_pipeline_transform(df):
    num = ['float64','int64']
    num_attr = df.select_dtypes(include = num)
    
    num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('scaler', MinMaxScaler(feature_range=(0,1)))
    ])
    return num_attr, num_pipeline

def pipeline_transf(df):
    cat_attr = list(df.select_dtypes(include=['object']).columns)
    num_attr, num_pipeline = numeric_pipeline_transform(df)
    complete_pipeline = ColumnTransformer([
        ('num', num_pipeline, list(num_attr)),
        ('cat', OneHotEncoder(handle_unknown="ignore"), cat_attr)
    ])
    proc_data = complete_pipeline.fit_transform(df)
    return proc_data
```

```python
df_train.head(2)
```

```python
label = df_train['y'].copy()
df_train.drop(columns = 'y', inplace = True)
proc_df = pipeline_transf(df_train)
```

```python
df = pd.DataFrame(proc_df, columns = ['x'+str(i) for i in range(proc_df.shape[1])])
```

```python
df = pd.concat([df, label], axis = 1)
df.head(2)
```

```python
split = StratifiedShuffleSplit(n_splits = 1, test_size = 0.2, random_state = 94)
for train_index, test_index in split.split(df, df['y']):
    str_train = df.loc[train_index]
    str_val = df.loc[test_index]
```

```python
str_train.y.value_counts(1)
```

```python
str_val.y.value_counts(1)
```

```python
train = str_train.drop(columns = ['y'])
train_label = str_train['y'].copy()

test = str_val.drop(columns = ['y'])
test_label = str_val['y'].copy()
```

```python
train.shape,train_label.shape, test.shape, test_label.shape
```

# 5. MODEL

```python
%%time
######## GBC oversamplig ########  
SMOTE = SMOTE()

train_SMOTE, train_label_SMOTE = SMOTE.fit_resample(train, train_label)


model = ensemble.GradientBoostingClassifier()
model.fit(train_SMOTE, train_label_SMOTE)
predictions = model.predict(test)
predicted_prob = model.predict_proba(test)[:,1]
print('*'*50)
print('matriz de confusion')
print(pd.crosstab(test_label, predictions))
print('*'*50)

accuracy = metrics.accuracy_score(test_label, predictions)
auc = metrics.roc_auc_score(test_label, predicted_prob)
print('*'*50)
print("Accuracy (overall correct predictions):",  round(accuracy,2))
print("Auc:", round(auc,2))
print('*'*50)    
print('precision: {}'.format(metrics.precision_score(test_label, predictions, average=None)))
print('recall: {}'.format(metrics.recall_score(test_label, predictions, average=None)))
print('f1 score: {}'.format(metrics.f1_score(test_label, predictions, average=None)))
```

```python
%%time
######## GBC undersamplig ########  

undersample = RandomUnderSampler(sampling_strategy='majority')

train_under, train_label_under = undersample.fit_resample(train, train_label)


model = ensemble.GradientBoostingClassifier()
model.fit(train_under, train_label_under)
predictions = model.predict(test)
predicted_prob = model.predict_proba(test)[:,1]
print('*'*50)
print('matriz de confusion')
print(pd.crosstab(test_label, predictions))
print('*'*50)

accuracy = metrics.accuracy_score(test_label, predictions)
auc = metrics.roc_auc_score(test_label, predicted_prob)
print('*'*50)
print("Accuracy (overall correct predictions):",  round(accuracy,2))
print("Auc:", round(auc,2))
print('*'*50)    
print('precision: {}'.format(metrics.precision_score(test_label, predictions, average=None)))
print('recall: {}'.format(metrics.recall_score(test_label, predictions, average=None)))
print('f1 score: {}'.format(metrics.f1_score(test_label, predictions, average=None)))
```

# 6. TUNE

```python
%%time
######## GBC undersamplig ########  
undersample = RandomUnderSampler(sampling_strategy='majority')
train_under, train_label_under = undersample.fit_resample(train, train_label)

model = ensemble.GradientBoostingClassifier()
model.fit(train_under, train_label_under)
param_dic = {'learning_rate':[0.15,0.1,0.05,0.01,0.005,0.001],
'n_estimators':[100,250,500,750,1000,1250,1500,1750],  
'max_depth':[2,3,4,5,6,7],   
'min_samples_split':[2,4,6,8,10,20,40,60,100],    
'min_samples_leaf':[1,3,5,7,9],     
'max_features':[2,3,4,5,6,7],     
'subsample':[0.7,0.75,0.8,0.85,0.9,0.95,1]}  
## random search
random_search = model_selection.RandomizedSearchCV(model, 
       param_distributions=param_dic, n_iter=100, 
       scoring="accuracy").fit(train_under, train_label_under)
print("Best Model parameters:", random_search.best_params_)
print("Best Model mean accuracy:", random_search.best_score_)
model = random_search.best_estimator_

predictions = model.predict(test)
predicted_prob = model.predict_proba(test)[:,1]
print('*'*50)
print('matriz de confusion')
print(pd.crosstab(test_label, predictions))
print('*'*50)

accuracy = metrics.accuracy_score(test_label, predictions)
auc = metrics.roc_auc_score(test_label, predicted_prob)
print('*'*50)
print("Accuracy (overall correct predictions):",  round(accuracy,2))
print("Auc:", round(auc,2))
print('*'*50)    
print('precision: {}'.format(metrics.precision_score(test_label, predictions, average=None)))
print('recall: {}'.format(metrics.recall_score(test_label, predictions, average=None)))
print('f1 score: {}'.format(metrics.f1_score(test_label, predictions, average=None)))
```

# 8. PROSPECT

```python
df_pred = pd.read_csv('Data_test.csv')
print(df_pred.shape)
print(df_pred.CLIENTE.nunique())
df_pred.head(2)
```

```python
def final_prospect(df, final_model):
    df2 = cat_num_type(df.drop(columns = ['CLIENTE']))
    proc_pred = pipeline_transf(df2)
    fill_dim = np.zeros((proc_pred.shape[0],195 - proc_pred.shape[1]))
    proc_pred = np.hstack((proc_pred,fill_dim))
    
    with open(final_model, 'rb') as file_input:
        model = pickle.load(file_input)
    
    y_pred = model.predict(proc_pred)
    df['y'] = y_pred
    
    return df
```

```python
df = final_prospect(df_pred, final_model = 'final_model.pkl')
```

# Conclusiones Finales
1. Se considero como categóricas aquellas variables con menos de 20 valores únicos, esto debido a que no se tiene mayor descripción de cada uno de los features

2. Se implemento el EDA con la librería sweetviz que permite realizar gráficos de manera automática para una gran cantidad de variables en formato html.

3. El data engineering, imputación y construcción de dummies se realizó con pipeline para mayor facilidad.

4. Se probó 3 algoritmos principalmente: Catboost, Random Forest, y Gradient Boosting. En este último se implementó oversamplig y undersamplig para un mejor resultado. Al final nos quedamos con undersamplig por ser más económico en término de recursos y que nos da mejores resultados.

5. Se usó RandomSearch para encontrar la combinación de hiperparámetros óptima.

6. El modelo final es un Gradient Boosting usando undersampling. Tenemos un precisión de 8% y recall de 75%, un f1 score de 0.14.

7. Como nosotros estamos tratando de predecir la fuga de clientes, el indicador más útil es el recall. Me conviene predecir de los que realmente se irán, la mayor cantidad de clientes, 75% de ellos para ser exacto.

8. Como oportunidad de mejora se puede adicionalmente implementar una función de pareto para reducir la cantidad de clases en las variables categóricas.

9. El negocio puede usar el modelo en el caso de fuga para priorizar estos leads en las gestiones y darle descuentos o ofertas diferenciadas acorde a su perfil.
